### 4- Separación del dataset en entrenamiento y validación.

```python
X = data.copy(deep=True)
y = X.pop('RECARGA')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, random_state = 30, stratify=y)
```

### 5- Búsqueda de outliers.

```python
columnas = ['EDAD', 'RECARGA_ANTERIOR', 'ANTIG', 'SEGUNDOS_TRAFICO_ACTUAL', 'SEGUNDOS_TRAFICO_ANTERIOR', 'MB_TRAFICO_ACTUAL', 'MB_TRAFICO_ANTERIOR']

def es_outlier(df, columna, umbral=1.5):
  q1 = df[columna].quantile(0.25)
  q3 = df[columna].quantile(0.75)
  iqr = q3 - q1
  return (df[columna]<q1-umbral*iqr)|(df[columna]>q3+umbral*iqr) 

contador = 0

print('Cantidad de outliers por columna:')
for i in columnas:
  x = X_train.loc[es_outlier(X_train, i)].shape[0]
  contador += 1
  print(f'{contador}- {i}: {x}')
```

### 6- Imputación de outliers.

```python
def imputador_outlier(data,variables):
    
    outliers = data.copy()
    
    for i in variables:        
        Q1 = outliers[i].quantile(0.25)
        Q3 = outliers[i].quantile(0.75)
        IQR = Q3 - Q1
        valor_menor = (Q1 - 1.5 * IQR)
        valor_mayor = (Q3 + 1.5 * IQR)
        
        outliers.loc[outliers[i] < valor_menor,i] = valor_menor
        outliers.loc[outliers[i] > valor_mayor,i] = valor_mayor
        
    return outliers 

#nuevos_datos = data.select_dtypes(include='float')
#variables2 = nuevos_datos.columns

X_train_mod = imputador_outlier(X_train,columnas)  
X_train_mod
```

```python

# Número de gráficos por fila
graficos_por_fila = 4
# Número total de gráficos
total_graficos = 7

# Calcular el número de filas necesarias
filas = (total_graficos + graficos_por_fila - 1) // graficos_por_fila

fig, axes = plt.subplots(filas, graficos_por_fila, figsize=(12, 5 * filas))

for i in range(total_graficos):
    fila = i // graficos_por_fila
    columna = i % graficos_por_fila
    sns.boxplot(X_train_mod[columnas[i]], ax=axes[fila, columna])
    axes[fila, columna].set_title(f'Gráfico de caja #{i+1}')

# Ajustar el layout para que no se solapen los títulos
plt.tight_layout()
plt.show()
```

### 7- Normalización de columnas con valores no binarios.

```python

from sklearn.preprocessing import MinMaxScaler


scaler_minmax = MinMaxScaler()

scaler_minmax.fit(X_train_mod[columnas])

X_train_mod[columnas] = scaler_minmax.transform(X_train_mod[columnas])
X_val[columnas] = scaler_minmax.transform(X_val[columnas])
```

```python
X_train_mod
```

### 8- Eliminación de columnas innecesarias.

#### La columna 'ID' no aporta valor como variable al modelo que entrenaremos ya que solo es un identificador de registros; por lo que se procede a eliminarla del dataset para entrenamiento. Por otro lado, en las columnas 'TECNOLOGIA_TRAFICO_3G' y 'TECNOLOGIA_TRAFICO_4G', basta con mantener solo una pues estan relacionadas directamente.

```python
X_train_scaled = X_train_mod.drop(columns=['ID', 'TECNOLOGIA_TRAFICO_3G'])
X_val_scaled = X_val.drop(columns=['ID', 'TECNOLOGIA_TRAFICO_3G'])
X_train_scaled
```

### 9- Verificación del balance de la variable objetivo.

```python
y_train.value_counts(dropna=False, normalize=True)
```

#### Al tener una proporción de 78/22, se optará por balancear los datos con la técnica de Random Oversampling.

### 10- Balanceo con Random Oversampling.

```python
from imblearn.over_sampling import RandomOverSampler

oversample = RandomOverSampler(sampling_strategy=0.50)

X_train_over, y_train_over = oversample.fit_resample(X_train_scaled, y_train)
```

```python
y_train_over.value_counts(dropna=False, normalize=True)
```

### 11- Entrenamiento de modelos de machine learning para clasificación.

Como parámetro de comparación para medir el desempeño de los modelos, se decidió por el RECALL ya que para este caso se busca generar que la gente realize recargas, es decir que se busca identificar a la mayor cantidad posible de clientes que no recargan. Para ello, el recall es la mejor opción ya que mínimiza los falsos negativos y máximiza la identificacion de clientes que no recargan.

```python
from sklearn.metrics import recall_score
```

#### a) Árboles de decisión.

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(max_depth=10)
tree.fit(X_train_over, y_train_over)
```

```python
y_pred_tree = tree.predict(X_train_over)
recall_score(y_train_over, y_pred_tree)
```

##### Validación del modelo.

```python
y_pred_tree_val = tree.predict(X_val_scaled)
```

```python
recall_tree = recall_score(y_val, y_pred_tree_val)
recall_tree
```

#### b) Random forest.

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=100, max_depth=10, min_samples_split=5, min_samples_leaf=2)
forest.fit(X_train_over, y_train_over)
```

```python
y_pred_forest = forest.predict(X_train_over)
```

```python
recall_score(y_train_over, y_pred_forest)
```

##### Validación del modelo.

```python
y_pred_forest_val = forest.predict(X_val_scaled)
```

```python
recall_forest = recall_score(y_val, y_pred_forest_val)
recall_forest
```

#### c) Naive bayes.

```python
from sklearn.naive_bayes import GaussianNB

clf = GaussianNB()
clf.fit(X_train_over, y_train_over)
```

```python
y_pred_naive = clf.predict(X_train_over)
```

```python
recall_score(y_train_over, y_pred_naive)
```

##### Validación del modelo.

```python
y_predic_naive_val = clf.predict(X_val_scaled)
```

```python
recall_nb = recall_score(y_val, y_predic_naive_val)
recall_nb
```

#### d) Regresión logística.

```python
from sklearn.linear_model import LogisticRegression

logr = LogisticRegression(max_iter=1000)
logr.fit(X_train_over, y_train_over)
```

```python
y_pred_logr = logr.predict(X_train_over)
```

```python
recall_score(y_train_over, y_pred_logr)
```

##### Validación del modelo.

```python
y_predic_logr_val = logr.predict(X_val_scaled)
```

```python
recall_logr = recall_score(y_val, y_predic_logr_val)
recall_logr
```

#### Comparación del desempeño de los modelos.

```python
print(f'Recall - Árbol de decisión: {recall_tree:.4f}')
print(f'Recall - Random forest: {recall_forest:.4f}')
print(f'Recall - Naive Bayes: {recall_nb:.4f}')
print(f'Recall - Regresión logística: {recall_logr:.4f}')
```
